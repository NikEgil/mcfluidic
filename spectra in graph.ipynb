```python
s0,s1,s2,s3,s4=[],[],[],[],[]

s0=pd.read_excel('spec0.xlsx').to_numpy()
s1=pd.read_excel('spec1.xlsx').to_numpy()
s2=pd.read_excel('spec2.xlsx').to_numpy()
s3=pd.read_excel('spec3.xlsx').to_numpy()
s4=pd.read_excel('spec4.xlsx').to_numpy()

mas=[s0,s1,s2,s3,s4]
```

```python
plt.figure(figsize=(12,8))
x=np.arange(0,875)
color=['tab:red','tab:green','tab:blue','tab:orange','tab:gray']
color=['red','green','blue','orange','gray']
tp=[0,1,2,3,4]
for i in (tp):
        y1=mas[i][np.argmax(np.sum(mas[i], axis=1))]
        y2=mas[i][np.argmin(np.sum(mas[i], axis=1))]
        plt.fill_between(x, y1,y2, alpha=0.1,lw=2,color=color[i])
        plt.plot(x,y1,lw=1,color=color[i],label=i)
        plt.plot(x,y2,lw=1,color=color[i])

plt.legend()
```
